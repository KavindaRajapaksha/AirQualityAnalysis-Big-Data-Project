## Data Preprocessing

```python
import pandas as pd
```

```python
df=pd.read_csv('station_hour.csv')
```

```python
df.head()
```

```python
df.tail(20)
```

```python
print(df.columns)
```

```python
df=df[(df['PM2.5'].notna())&(df['AQI_Bucket'].notna())]
df = df[df['PM2.5'] != 'NA']
df = df[df['AQI_Bucket'] != 'NA']
```

```python
df.describe()
```

```python
df.info()
```

```python
df.shape
```

```python
df = df[['StationId', 'Datetime', 'PM2.5', 'AQI_Bucket']]
```

```python
df.shape
```

```python
df.info()
```

```python
df['PM2.5']=pd.to_numeric(df['PM2.5'],errors='coerce')
```

```python
df = df.dropna(subset=['PM2.5'])
```
